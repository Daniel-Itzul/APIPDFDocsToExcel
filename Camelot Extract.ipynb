```python
import pandas as pd
import camelot as cmt
```

```python
tables = cmt.read_pdf('File_Specifications_for_SA_028+_standard.pdf',pages='25-44')
```

```python
tables[0].df
```

```python
writer = pd.ExcelWriter('0900 Detail in scope.xlsx',engine='xlsxwriter')
workbook=writer.book
worksheet=workbook.add_worksheet('rawData')
writer.sheets['rawData'] = worksheet
initRow = 0
tableIndex = 0
for table in tables:
    tempDf = tables[tableIndex].df
    tempDf.to_excel(writer,sheet_name ='rawData',startrow=initRow , startcol=0, index=False)
    initRow = initRow + table.df.shape[0] + 2
    tableIndex = tableIndex + 1
writer.save()
```
